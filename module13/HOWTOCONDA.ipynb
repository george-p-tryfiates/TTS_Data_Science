#### *Installation*

*From Anaconda*:

Launch the Anaconda Terminal.

Run this command in Anaconda: conda install -c r rstudio

Run this command in Anaconda to download Anaconda Navigator: conda install anaconda-navigator

Open Anaconda navigator and go to the Environments tab. 

Click the buttom below your Environments called Create.

Select both Python and R and click Ok.

After running, you can switch to the Home tab and change the environment to your joint programming language environment and you can launch Jupyter Notebook. From this, you can create notebooks in either language.

*From CRAN*:

Select a mirror and then download from that page the appropriate installer for your OS: https://cran.r-project.org/mirrors.html

### Install R Studio 

    What is R Studio?
An IDE for R. You can create R Markdown files and R Notebooks for doing project write-ups with code chunks or just annotating code. Each of the files allow you to insert code chunks into a text block and then "knit" the projects together to publish them to Word, PDF, or HTML files.

    What is a script?
A script is an R code file.

    What is a notebook?
A notebook allows the assembly of script and text into a file.

    Post a picture of the R studio screen and explain the different parts like the console


We have a few basic windows in the RStudio IDE: the editor (top left), the environment (top right), the console (bottom left),and the browser (bottom right). 

The editor is where open projects exist as tabs. We can edit / run / interact with the code / text here.

The environment window shows all existing objects in the environment. These can be clicked to show their properties / attributes. There are also history and connection tabs to show all previously run commands and online connections.

The console is a command terminal for R. You can click the up arrow key to access previously run commands. The terminal tab is the Window's terminal and I've never used it in the IDE or the other tabs so for now they will remain unused.

The browser allows the browsing of files, the viewing of plots, searching/managing packages, searching documentation under help, and other object viewing.

For more info check out: https://grunwaldlab.github.io/analysis_of_microbiome_community_data_in_r/00--intro_to_rstudio.html
